### EDA

```python
# plot correlation matrix with fig size 10,10
plt.figure(figsize=(10,10))
corr = sal.corr()
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns)
```

```python
# plot pairplot
sns.pairplot(sal)
```

#### Univariate EDA

```python
# plot a histogram for LCURRENT
plt.figure(figsize=(10,10))
plt.hist(sal['LCURRENT'])
plt.title('Current Salary')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
# plot a histogram for LSTART
plt.figure(figsize=(10,10))
plt.hist(sal['LSTART'])
plt.title('Starting Salary')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```
